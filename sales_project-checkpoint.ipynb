```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a sample sales dataset
data = {
    'Date': pd.date_range(start='2024-01-01', periods=15),
    'Product': ['A', 'B', 'C', 'D', 'E'] * 3,
    'Category': ['Electronics', 'Clothing', 'Grocery', 'Electronics', 'Clothing'] * 3,
    'Units_Sold': [10, 20, 15, 40, 35, 25, 30, 45, 50, 20, 15, 10, 25, 40, 30],
    'Unit_Price': [100, 50, 20, 200, 80, 110, 55, 25, 210, 85, 105, 60, 30, 220, 90]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Step 2: Data Cleaning
# Add a 'Total_Sales' column
df['Total_Sales'] = df['Units_Sold'] * df['Unit_Price']

# Step 3: Analysis
# a. Total sales by product
sales_by_product = df.groupby('Product')['Total_Sales'].sum().reset_index()

# b. Total sales by category
sales_by_category = df.groupby('Category')['Total_Sales'].sum().reset_index()

# c. Total units sold by date
units_by_date = df.groupby('Date')['Units_Sold'].sum().reset_index()

# Step 4: Visualization

# a. Bar plot for total sales by product
plt.figure(figsize=(8, 5))
sns.barplot(x='Product', y='Total_Sales', data=sales_by_product, palette='viridis')
plt.title('Total Sales by Product', fontsize=16)
plt.ylabel('Total Sales', fontsize=12)
plt.xlabel('Product', fontsize=12)
plt.show()

# b. Pie chart for total sales by category
plt.figure(figsize=(6, 6))
plt.pie(sales_by_category['Total_Sales'], labels=sales_by_category['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Category', fontsize=16)
plt.show()

# c. Line plot for units sold over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Units_Sold', data=units_by_date, marker='o', color='green')
plt.title('Units Sold Over Time', fontsize=16)
plt.ylabel('Units Sold', fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.xticks(rotation=45)
plt.grid(visible=True)
plt.show()

# Step 5: Summary
print("Summary of Analysis:")
print("-" * 30)
print("Total Sales by Product:\n", sales_by_product)
print("\nTotal Sales by Category:\n", sales_by_category)
print("\nUnits Sold Over Time:\n", units_by_date)
```
